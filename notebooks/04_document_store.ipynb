```python
from rag.embeddings import create_embedder
from rag.retrieval.vector_store import DocumentStore
from rag.config import settings

embedder = create_embedder(settings)

store = DocumentStore(embedder, index_name='flat_ip')
store
```

```python
from datasets import load_dataset

ds = load_dataset("rag-datasets/rag-mini-bioasq", "text-corpus")['passages']
ds
```

```python
iter(ds['passage'])
```

```python
ds['passage'][:100]
```

```python
store.embedder.embed_batch(['Dopamine sources in human']).shape
```

```python
store.add_documents(ds['passage'], verbose=True)
```

```python
store.search("Where dopamine is secreted?", k=5)
```

```python
store.save('bioasq-mini')
```

```python
store2 = DocumentStore(embedder, index_name='flat_ip')
store2.load('bioasq-mini')
```

```python
store2.search("What is dopamine?", k=5)
```
